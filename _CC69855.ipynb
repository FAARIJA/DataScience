```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
iris = sns.load_dataset('iris')
```

```python
# Display the first few rows
print(iris.head())


```

```python

# Display data types and non-null counts
print(iris.info())
```

```python
# Summary statistics
print(iris.describe())
```

```python
# Set the aesthetic style of the plots
sns.set(style='whitegrid')
```

```python
# Histogram
plt.figure(figsize=(12, 6))
for i, column in enumerate(iris.columns[:-1]):
    plt.subplot(2, 2, i + 1)
    sns.histplot(iris[column], bins=15, kde=True)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()
```

```python
# Pairplot to see pairwise relationships
sns.pairplot(iris, hue='species', palette='Set1')
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', style='species', palette='Set1')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
# Box plot for Sepal Length across Species
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris, x='species', y='sepal_length', palette='Set2')
plt.title('Box plot of Sepal Length by Species')
plt.show()
```

```python
# Correlation matrix
# Exclude the 'species' column before calculating correlation
correlation = iris.drop(columns=['species']).corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```
